```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images_scaled = train_images / 255.0
test_images_scaled = test_images /255.0
```

## <span style="color:red">Hidden Layer Function</span>

```python
def get_model(hidden_layer=1):
    layers = []
    layer_neurons = 28*28
    layers.append(keras.Input(shape=(28, 28)))
    layers.append(keras.layers.Flatten())
    for i in range(hidden_layer):
        layers.append(keras.layers.Dense(500, activation='relu'),)
        
    layers.append(keras.layers.Dense(10, activation='sigmoid'))

    model = keras.Sequential(layers)

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python
model = get_model(3)
model.fit(train_images_scaled, train_labels, epochs=5)
```

```python
model.predict(test_images_scaled)[2]
```

## <span style="color:red">Performance Test</span>

```python
tf.config.experimental.list_physical_devices() 
```

```python
import time
start_time = time.time()
with tf.device('/CPU:0'):
    cpu_model = get_model(1)
    cpu_model.fit(train_images_scaled, train_labels, epochs=5)
end_time = time.time()
print(f"Total training time for 1 hidden layer on CPU: {end_time - start_time:.2f} seconds")
```

```python
import time
start_time = time.time()
with tf.device('/CPU:0'):
    cpu_model = get_model(5)
    cpu_model.fit(train_images_scaled, train_labels, epochs=5)
end_time = time.time()
print(f"Total training time for 5 hidden layer on CPU: {end_time - start_time:.2f} seconds")
```
