```python
# USER Behaviour dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('user_behavior_dataset.csv')
```

```python
# inspect data
df.head(5)
```

```python
# Get all columns name
df.columns
```

```python
# Rename columns
df.rename(columns={'Device Model':'Model', 'Operating System':'OS',
       'App Usage Time (min/day)':'App Usage', 'Screen On Time (hours/day)':'Screen Time',
       'Battery Drain (mAh/day)':'Battery Drain', 'Number of Apps Installed':"no of app",
       'Data Usage (MB/day)':'Data Usage'}, inplace=True)
```

```python
df.columns
```

```python
plt.figure(figsize=(5, 2.5));
plt.scatter(data=df, x='no of app', y='Screen Time', 
            s=5, c='Data Usage', cmap='Set3');
plt.xlabel('No of App Installed');
plt.ylabel('Screen Time');
plt.show;
```

```python
plt.figure(figsize=(3, 2));
plt.bar(df['OS'], df['Screen Time']);
```

```python
plt.figure(figsize=(3, 2));
plt.scatter(data=df, x='Age', y='Screen Time',s=5);
```

```python
### Find the screen time based on age
plt.figure(figsize=(7,5))
plt.scatter(data=df, x='Age', y='Screen Time',s=5)
plt.xlabel('Age')
plt.ylabel('Screen Time')
plt.title('Screen Time based on Age')
plt.show()
```
