# Manipulating Index

```python
gym.set_index("Weight (kg)")
```

```python
gym.BMI<=25 
```

```python
gym.loc[gym.Workout_Type=='Yoga']
```

That means among 973 people 239 do Yoga

```python
gym.loc[(gym.Workout_Type=='Yoga') & (gym.BMI<=25)]
```

Among 239 people who do Yoga, 135 people has BMI less or equal to 25 which is standard BMI level

```python
gym.loc[(gym.Workout_Type=='Yoga') & (gym.BMI<=25) & (gym.BMI>=20)]
```

68 people who does Yoga has a standard BMI (20<=BMI<=25)
