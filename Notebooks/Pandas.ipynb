```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell=="not available":
        return None
    return cell
def convert_revenue_cell(cell):
    if cell==-1:
        return None
    return cell

df=pd.read_excel("stock_data.xlsx", "Sheet1", converters={
    'people':convert_people_cell,
    'eps':convert_eps_cell,
    'revenue':convert_revenue_cell
})
df
```

# Basic Merge Using a Dataframe Column

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san francisco"],
    "humidity": [65,68,71],
})
df2
```

# Types of DataBase Joins

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator="true")
df3
```

# Suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=("_left","_right"))
df3
```

# Cross tabulation or Crosstab

```python
pd.crosstab(df.Sex,df.Handedness)
```
