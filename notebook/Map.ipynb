```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a GeoJSON or shapefile of Dhaka's boundaries
# Example: Replace 'dhaka_boundaries.geojson' with your file
gdf = gpd.read_file('dhaka_boundaries.geojson')

# Plot the area
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='red', alpha=0.4)  # Set color and transparency
plt.title('Dhaka Marked Area')
plt.show()
```
