```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
    
```

```python
digits.target[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, train_size=0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
model.predict(digits.data[0:5])
```

**Confusion Matrix --> to find wrong prediction of the model** 

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

#### This Matrix means:
- at (0,0) 22 times predicted=0=result
- at (6,6) 26 times predicted=6=result
- at (3,7) 1 times predicted=3 but result=7
- at (8,1) 1 times predicted=8 but result=1
