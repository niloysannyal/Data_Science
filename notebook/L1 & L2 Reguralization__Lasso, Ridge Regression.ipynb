```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Melbourne_housing_Full.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = df[['Suburb','Address','Rooms','Type','Method','SellerG','Distance','Regionname','Propertycount','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.head()
```

```python
print(df.dtypes.value_counts())
```

```python
print(df.select_dtypes(include=['object']).nunique())
```

```python
low_cardinality_cols = ['Type', 'Method', 'Regionname']
df = pd.get_dummies(df, columns=low_cardinality_cols, drop_first=True)
```

```python
high_cardinality_cols = ['Suburb', 'SellerG', 'Address']

for col in high_cardinality_cols:
    df[col] = df[col].map(df[col].value_counts())
```

```python
df.shape
```

```python
df[44:55]
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
Ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
Ridge_reg.fit(X_train, y_train)
```

```python
Ridge_reg.score(X_test, y_test)
```
