```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

# Manual Approach

```python
encoders = pd.get_dummies(df[['Car Model']])
encoders
```

```python
merged = pd.concat([df,encoders],axis='columns')
merged
```

```python
final_table = merged.drop(['Car Model','Car Model_Audi A5'], axis='columns')
final_table
```

```python
X = final_table.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = final_table[['Sell Price($)']]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X,y)
```

# Second Approach

```python
df
```

```python
X = df.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = df[['Sell Price($)']]
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer([
    ('ohe',OneHotEncoder(),[0])],
    remainder = "passthrough"
)
```

```python
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
final_X = transformed_X[:,1:]
final_X
```

```python
y
```

```python
model2 = LinearRegression()
```

```python
model2.fit(final_X,y)
```

```python
model2.predict([[0,1,45000,4]])
```

```python
model2.predict([[1,0,86000,7]])
```

```python
model2.score(final_X,y)
```
