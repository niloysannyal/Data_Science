```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies_sentiment_data.csv")
df.head()
```

```python
df.shape
```

```python
df.sentiment.value_counts()
```

```python
df['positive'] = df['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
```

```python
df.head()
```

```python
X = df['review']
y = df['positive']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
type(X_train)
```

```python
type(X_train.values)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
v.vocabulary_
```

```python
len(v.vocabulary_)
```

```python
dir(v)
```

```python
v.get_feature_names_out()[1000:1050]
```

```python
X_train_np = X_train_cv.toarray()
X_train_np[0]
```

### Model Building

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=500,
    max_depth=None,
    min_samples_split=5,
    max_features='sqrt',       
    oob_score=True,
    n_jobs=-1
)

model.fit(X_train_cv, y_train)
```

```python
X_train_cv.shape
```

```python
X_test_cv = v.transform(X_test)
X_test_cv.shape
```

```python
model.score(X_test_cv, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty='l2',        # standard regularization
    C=1.0,               # can tune: [0.01, 0.1, 1, 10]
    solver='saga',       # handles large sparse data well
    max_iter=1000,       # enough for convergence
    n_jobs=-1,           
    random_state=42
)

model.fit(X_train_cv, y_train)
```

```python
model.score(X_test_cv, y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```
